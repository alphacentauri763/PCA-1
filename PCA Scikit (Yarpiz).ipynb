```python
import numpy as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pyforest
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
Y = iris.target
```

```python
X.shape
```

```python
Y.shape
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
pca.components_
```

```python
Z = pca.transform(X)
```

```python
Z.shape
```

```python
plt.scatter(Z[:,0], Z[:,1], c=Y)
```
